### 2. From string to float

```python
# distance

raw['distance'] = [i.replace(',','.') for i in list(raw.distance)]
raw['distance'] = [float(i) for i in list(raw.distance)]


#consume

raw['consume'] = [i.replace(',','.') for i in list(raw.consume)]
raw['consume'] = [float(i) for i in list(raw.consume)]

#temp inside

tempind = []

for i in list(raw.temp_inside):
    try:
        tempind.append(float(i.replace(',','.')))
    except:
        tempind.append(np.nan)

raw['temp_inside'] = tempind
```

```python
rawfloat = raw
```

```python
rawfloat.describe()
```

### 4. Linear regression

Preparing variables

```python
from scipy.stats import linregress

import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# y=mx+b , mpg=m*acceleration + b

x=sm.add_constant(raw_fdg[[
    'rain',
    'snow',
    'rain',
    'temp_outside',
    'temp_inside',
    'distance',
    'speed',
    'AC',
    'E10'
]].apply(lambda x: x.fillna(x.mean()))

)

y=raw_fdg['consume']


modelo=sm.OLS(y, x).fit()
modelo.summary()
```
