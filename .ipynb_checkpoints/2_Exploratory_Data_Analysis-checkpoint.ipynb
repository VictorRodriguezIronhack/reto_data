```python
import pandas as pd
import numpy as np
import seaborn as sns
import pylab as plt
```

```python
data = pd.read_csv('data/clean_data.csv')
```

```python
data.head()
```

La columna objetivo es **'gas_type'** ya que lo que queremos saber es cuál es el combustible que ofrece mejores resultados.

Construiremos algunos gráficos que nos permitan entender un poco mejor la situación.

```python
data.gas_type.value_counts()

# Existe un desbalance en los datos, lo cual puede provocar desajustes a futuro (desarrollo de modelos).
```

```python
# Matplotlib inline para visualizar los gráficos de Matplotlib
%matplotlib inline
%config Inlinebackend.figure_format= 'retina'

# Configuración para setear y que todas las fig de Seaborn salgan por defecto con este tamaño
# Se puede especificar el tamaño de cada figura
sns.set_context("poster")
sns.set(rc={"figure.figsize": (12.,6.)})
sns.set_style("whitegrid")
```

```python
ax = sns.boxplot(x="gas_type", y="consume", data=data)
ax = sns.swarmplot(x="gas_type", y="consume", data=data, color=".25")
```

```python
sns.kdeplot(x=data.consume, hue=data.gas_type)
```

```python
graf = sns.histplot(x=data.consume, kde=True, hue=data.gas_type)
graf.axvline(x=data.consume.mean(), c="red", label="mean")
graf.axvline(x=data.consume.median(), c="green", label="median")
```

```python
sns.scatterplot(x="distance", y="consume", hue="gas_type",data=data);
```

```python
sns.scatterplot(x="speed", y="consume", hue="gas_type",data=data);
```

```python
mascara = np.triu(np.ones_like(data.corr(), dtype=bool)) # generamos la máscara para la triangular superior

color_map = sns.diverging_palette(220, 20, as_cmap=True) # Paleta de colores

plt.figure(figsize=(14,14)) 

sns.heatmap(data.corr(),  
            mask = mascara,
            cmap=color_map,
            square=True, #que los datos se vean como cuadrados
            linewidth=0.5, #ancho de línea
            vmax=1,
            vmin=-1,
            cbar_kws={"shrink": .7
                     },# barra lateral
            annot=True);
```

- Hipótesis: "Los vehículos que utilizan combustible SP98 a priori muestra un consumo menor a los que utilizan E10."
