```python
df_a = df[df.distance<=50].reset_index(drop=True)
```

# Feature engineering

```python
df_a['gasoline'] = df_a.gas_type.apply(lambda x: 0 if x =='E10' else 1)
```

## Type of drives

```python
X =  df_a[['distance','speed']]
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
import matplotlib.pyplot as plt
```

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(
    model, k=(2,8), metric='calinski_harabasz', timings=False
)

visualizer.fit(X_scaled)        # Fit the data to the visualizer
visualizer.show()
```

```python
k_means = KMeans(n_clusters=5, 
                 random_state=0)

k_means.fit(X_scaled)

data = pd.DataFrame(X_scaled, columns=['distance', 'speed'])
data['cluster'] = k_means.labels_

sns.scatterplot(x = data.distance, y = data.speed, data=data, s= 100, hue=data.cluster)
plt.xlabel('distance', size=20)
plt.ylabel('speed', size=20)
# plt.scatter(data.PC1, data.PC2)
plt.show()
```

```python
data
```

```python
df_a
```

```python
df_a['type_drive'] = data['cluster'].copy()
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df_list = [df_a[df_a.gasoline == 0], df_a[df_a.gasoline == 1]]
```

```python
df_gas = df_list[0]
X = df_gas[['temp_outside','AC','rain','sun','type_drive']]
y = df_gas['consume']
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                y, 
                                                test_size=0.1, 
                                                random_state=46)    
```

```python
gb = GradientBoostingRegressor()

parameter_space = {'learning_rate': [0.01, 0.001],
                   'n_estimators': [2000,3000],
               'max_features': [1],
               'max_depth': [2,3],
               'min_samples_leaf': [7, 10,15]}

grid_search = GridSearchCV(gb,
                   param_grid=parameter_space,
                   verbose=1,
                    n_jobs= -1,
                   cv=5)

grid_search.fit(X_train, y_train)

gb_best = grid_search.best_estimator_
y_predict_test = gb_best.predict(X_test) 

score = mean_squared_error(y_test, y_predict_test)

params = (grid_search.best_params_)
```

```python
score
```

```python
params
```

```python
gb_best.fit(X,y)
```

```python
list_predictors = []
```

```python
list_predictors.append(gb_best)
```
